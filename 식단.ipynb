```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt                      
import seaborn as sns
import matplotlib as mpl
mpl.rcParams['font.family']='NanumGothic'

data=pd.read_csv("school_health_fixed.csv")
data=data[data['아침식사'].notna()]

# BMI 지수
conditions = [(data['BMI'] >= 25), (data['BMI'] < 25) & (data['BMI'] >= 23), (data['BMI'] < 23) & (data['BMI'] >= 18.5), (data['BMI'] < 18.5)]
choices = ['비만', '과체중', '정상', '저체중']
data.loc[:, 'BMI 지수'] = np.select(conditions, choices)

data['BMI 지수'] = pd.Categorical(data['BMI 지수'], categories=['저체중', '정상', '과체중' , '비만'], ordered=True)

plt.figure(figsize=(30, 7))
plt.subplot(1, 4, 1)
sns.histplot(data=data, x='우유유제품', discrete=True, kde=False)
plt.xticks([1, 2, 3, 4])
plt.title('혈압 분포 히스토그램')

plt.subplot(1, 4, 2)
sns.histplot(data=data, x='과일', discrete=True, kde=False)
plt.xticks([1, 2, 3, 4])
plt.title('혈압 분포 히스토그램')

plt.subplot(1, 4, 3)
sns.histplot(data=data, x='채소김치제외', discrete=True, kde=False)
plt.xticks([1, 2, 3, 4])
plt.title('혈압 분포 히스토그램')

plt.subplot(1, 4, 4)
sns.histplot(data=data, x='아침식사', discrete=True, kde=False)
plt.xticks([1, 2, 3, 4])
plt.title('혈압 분포 히스토그램')
```
